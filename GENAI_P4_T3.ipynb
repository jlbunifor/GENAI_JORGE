```python
import openai
from openai import OpenAI
apinumber='sua api aqui '
openai.api_key = apinumber
client = OpenAI(api_key=apinumber)
```

## text to speech

```python
meu_texto="Eu sou o Jorge Luiz. Professor de Ciencia de dados e Inteligencia Artificial"# esse texto sera transformado
response=client.audio.speech.create(model="gpt-4o-mini-tts",voice="coral",
                                                            input=meu_texto)
```

```python
# salvar o audio
with open("meu_audio.mp3", "wb") as f:
    f.write(response.content)
```

```python
# para ouvir o audio 
Audio(response.content)
```

## Speech to text

```python
with open("meu_audio.mp3", "rb") as audio_file:
    transcription = client.audio.transcriptions.create(
        model="whisper-1",               
        file=audio_file,                 
        response_format="text",         
        language="pt"                    
    )
```
